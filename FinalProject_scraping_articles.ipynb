```python
from bs4 import BeautifulSoup
import os
import requests
from datetime import datetime
import ast
import time


#def remove_non_ascii(text):
#    return ''.join([i if ord(i) < 128 else ' ' for i in text])

list=[]
for i in urls[0:549]:  #βάζω συγκεκριμένο αριθμό για τεστ
    dic = {}
    #url = base_url + str(i)
    print(i)
    page = requests.get(i, headers = headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    
    title = soup.select("h1")

    if (title):
        dic['title'] = soup.select("title")[0].text
    #else:
        #print("ARTICLE NO: ", link, " HAS NO TITLE")
        #dic['title'] = "NO TITLE"
    
        
    date = soup.select("div.field-item.even > p > em")
    if (date):
        dic['date'] = soup.select("div.field-item.even > p > em")[0].text
    #else:
        #print("ARTICLE NO: ", link, " HAS NO date")
        #dic['date'] = "NO date"
              
        
    
    article =  soup.select("div.field-item.even")
    if (article):
        dic['body'] = soup.select("div.field-item.even")[0].text
    #else:
        #print("ARTICLE NO: ", link, " HAS NO body")
        #dic['body'] = "NO body"  
    #article =  soup.select("#modal-ready")[0].text
    
    #article_text = [a.text for a in article]
    #dic['body'] = article
    
    dic['url'] = i

    
    list.append(dic)
    #print(dic)
    
    time.sleep(0.5) #ορίζω πόση ώρα θα περιμένει μέχρι να ξαναχτυπήσει τον σέρβερ για να πάρει το επόμενο
```

```python
df_list = pd.DataFrame(list)
df_list
```

```python
df_list.drop_duplicates(subset=['url'], keep='first', inplace = True)
df_list.reset_index(drop=True, inplace = True)
```
